```python
titles=pd.read_csv("Movie_Id_Titles")
titles.head()
```

```python
data=pd.merge(df,titles,on='item_id')
data.head()
```

```python
data.title.nunique()
```

```python
data.groupby('title')['rating'].count().sort_values(ascending=False).head()
```

```python
ratings=pd.DataFrame(data.groupby('title')['rating'].mean())
ratings.head()
```

```python
ratings['num of rating'] = pd.DataFrame(data.groupby('title')['rating'].count())
ratings.head(20)
```

moviemat=data.pivot_table(index='user_id',columns='title',values='rating')
moviemat.head()

```python
moviemat=data.pivot_table(index='user_id',columns='title',values='rating')
moviemat.head()
```

```python
print(moviemat.notnull())
```

```python
x=input()
#y=input()
x_user_ratings = moviemat[x]
#y_user_ratings = moviemat[y]
x_user_ratings.head()
#y_user_ratings.head()
```

```python
similar_to_x = moviemat.corrwith(x_user_ratings)
#similar_to_y = moviemat.corrwith(y_user_ratings)
```

```python
corr_x = pd.DataFrame(similar_to_x,columns=['Correlation'])
corr_x.dropna(inplace=True)
corr_x.head()
```

```python
corr_x.sort_values('Correlation',ascending=False).head(10)
```

```python
corr_x= corr_x.join(ratings['num of rating'])
corr_x.head()
```

```python
corr_x[corr_x['num of rating']>100].sort_values('Correlation',ascending=False).head()
```
